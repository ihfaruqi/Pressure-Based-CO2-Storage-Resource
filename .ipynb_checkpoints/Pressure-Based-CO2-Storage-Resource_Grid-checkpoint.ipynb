```python
# Constants for Reservoir and water depth (optional)
nresthick = float(input("Reservoir Thickness (m): ")) # default 10
nresntg = float(input("Reservoir N:G (fraction): ")) # default 0.5
nrespor = float(input("Reservoir Porosity (%): ")) # default 25
nwd = float(input("Water Depth (m): ")) # default 50
```

## CO2 Properties 

```python
# Crit Temp (K)
crittemp = 304.128
# Crit Pressure (Mpa)
critpressure = 7.377
# accentric factor
accfac = 0.228
# R (J/mol-K)
r = 8.314
# kappa
kappa = 0.37464 + 1.54226 * accfac - 0.26992 * accfac ** 2
# a-c 
aminc = 0.45724 * r**2 * crittemp**2 / critpressure
# b
b = 0.077796 * r * crittemp/critpressure
```

## Surface Water (seawater or lake) Properties

```python
# Surface water atmospheric pressure (Mpa)
nsurpres = 0.1

# Mean Annual Surface Water (Mudline) Temperature (degC). Forrest et al. (2007)
Ntemp = -0.04 * df['positive_water_depth'] + 24.44

# surface water density (g/cc)
def surface_water_density(Ntemp, Nsurwatsal, nsurpres):
    term1 = 1 + 1e-6 * (-80 * Ntemp - 3.3 * Ntemp ** 2 + 0.00175 * Ntemp ** 3 + 489 * nsurpres - 2 * Ntemp * nsurpres + 0.016 * Ntemp ** 2 * nsurpres - 1.3e-5 * Ntemp ** 3 * nsurpres - 0.333 * nsurpres ** 2 - 0.002 * Ntemp * nsurpres ** 2)
    term2 = Nsurwatsal * (0.668 + 0.44 * Nsurwatsal + 1e-6 * (300 * nsurpres - 2400 * nsurpres * Nsurwatsal + Ntemp * (80 + 3 * Ntemp - 3300 * Nsurwatsal - 13 * nsurpres + 47 * nsurpres * Nsurwatsal)))
    
    nsurwatden_gcc = term1 + term2
    return nsurwatden_gcc
nsurwatden_gcc = surface_water_density(Ntemp, Nsurwatsal, nsurpres)

# surface water density (kg/m3)
nsurwatden_kgm3 = nsurwatden_gcc*1000

# surface water pressure gradient (Mpa/km)
surwatpresgrad = nsurwatden_kgm3*9.8/1000
```

## Subsurface Brine Properties

```python
# Initialize depth and brinetemp
depthbml_m = list(range(0, 2001, 100))
brinetemp_c = [Ntemp + depthbml_m / 1000 * Ngeograd for depthbml_m in depthbml_m]

# Initialize empty lists for the new calculated values
brinepres_mpa = []
brineden_gcc = []
brineden_kgm3 = []

# Initialize nsurpres for the first row
F49 = nsurpres

# Initialize E49 for the first row (brineden_kgm3 for the first row)
E49 = 0  # Initialize with a value, since there's no previous E49 for the first row

# Loop through each row (assuming depthbml_m and brinetemp_c have the same length)
for i in range(len(depthbml_m)):
    if i == 0:
        # Calculate brinepres_mpa for the first row
        brinepres_mpa.append(nsurpres)
    else:
        # Calculate brinepres_mpa for the second and the rest of the rows
        B49 = depthbml_m[i - 1]
        B50 = depthbml_m[i]
        F50 = F49 + (B50 - B49) * E49 * 9.8 / 1000000
        brinepres_mpa.append(F50)
        F49 = F50  # Update F49 for the next iteration

    # Calculate brineden_gcc (same as before)
    C49 = brinetemp_c[i]
    brineden_gcc_formula = (
        (1 + 10 ** -6 * (-80 * C49 - 3.3 * C49 ** 2 + 0.00175 * C49 ** 3 + 489 * F49 - 2 * C49 * F49 + 0.016 * C49 ** 2 * F49 - 1.3 * 10 ** -5 * C49 ** 3 * F49 - 0.333 * F49 ** 2 - 0.002 * C49 * F49 ** 2))
        + Nresbrine * (0.668 + 0.44 * Nresbrine + 10 ** -6 * (300 * F49 - 2400 * F49 * Nresbrine + C49 * (80 + 3 * C49 - 3300 * Nresbrine - 13 * F49 + 47 * F49 * Nresbrine)))
    )
    brineden_gcc.append(brineden_gcc_formula)

    # Calculate brineden_kgm3 (same as before)
    brineden_kgm3_formula = brineden_gcc_formula * 1000
    brineden_kgm3.append(brineden_kgm3_formula)
    E49 = brineden_kgm3_formula  # Update E49 for the next iteration

# Now you have your new lists ready: brinepres_mpa, brineden_gcc, brineden_kgm3

# Subsurface hydrostatic pressure gradient (Mpa/km)
nsubhydpresgrad = brinepres_mpa[-1] / 2
```

## Sediment Density and Lithostatic Pressure

```python
# Calculate sandpor_pct using the given formula
sandpor_pct = [44 * math.exp(-depth / 2966) for depth in depthbml_m]

# Calculate shalepor_pct using the given formula
shalepor_pct = [62 * math.exp(-depth / 1472) if depth < 1680 else 33 * math.exp(-depth / 3299) for depth in depthbml_m]

# Calculate bulkden_kgm3 using the given formula
bulkden_kgm3 = [
    Noverburden * ((1 - sandpor / 100) * 2650 + sandpor / 100 * brineden) +
    (1 - Noverburden) * ((1 - shalepor / 100) * 2650 + shalepor / 100 * brineden)
    for sandpor, shalepor, brineden in zip(sandpor_pct, shalepor_pct, brineden_kgm3)
]

# Initialize lithpres_mpa with the first value (Nsurpres)
lithpres_mpa = [nsurpres]

# Calculate lithpres_mpa using the given formula
for depth_prev, depth_curr, bulkden_prev, bulkden_curr in zip(depthbml_m[:-1], depthbml_m[1:], bulkden_kgm3[:-1], bulkden_kgm3[1:]):
    lithpres_curr = lithpres_mpa[-1] + (depth_curr - depth_prev) * (bulkden_curr + bulkden_prev) / 2 * 9.8 / 1000000
    lithpres_mpa.append(lithpres_curr)

#Lithostatic Pressure Gradient (Mpa/km)
nlitpresgrad = lithpres_mpa[-1] / 2
```

# Reservoir Properties

```python
# Reservoir Injection Zone Thickness, Water Depth, Net Reservoir, Porosity
""" This one needs to be refined once 
the spreadsheet net res thickness and porosity available

Needs to be available: X_M, Y_M, positive_depth, true_thick, positive_water_depth, Net Reservoir Thickness (m)"""

def calculate_injection_zone_thickness(row):
    if pd.notna(row['true_thick']):
        return row['true_thick']
    else:
        return nresthick
df['Injection Zone Thickness (m)'] = df.apply(calculate_injection_zone_thickness, axis=1)

def calculate_midreservoir_depth(row):
    if pd.notna(row['MIDDEPTH_M']):
        return row['MIDDEPTH_M']
    else:
        return row['positive_depth'] + row['Injection Zone Thickness (m)'] / 2
df['Mid-Point Reservoir Depth (m)'] = df.apply(calculate_midreservoir_depth, axis=1)

def calculate_water_depth(row):
    if pd.notna(row['positive_water_depth']):
        return row['positive_water_depth']
    else:
        return nwd
df['Water Depth(m)'] = df.apply(calculate_water_depth, axis=1)

def calculate_netreservoir_thick(row):
    if pd.notna(row['NetRes_M']):
        return row['NetRes_M']
    else:
        return row['Injection Zone Thickness (m)'] * row['Mid_NTG']
df['Net Reservoir Thickness (m)'] = df.apply(calculate_netreservoir_thick, axis=1)

def calculate_porosity(row):
    depth = row['Mid-Point Reservoir Depth (m)']
    water_depth = row["Water Depth(m)"]
    if pd.notna(row['Phi%']):
        return row['Phi%']
    else:
        return "" if depth < water_depth else 44 * np.exp(-(depth-water_depth)/2966)

df['Reservoir Porosity (%)'] = df.apply(calculate_porosity, axis=1)
df['Reservoir Porosity (%)'] = pd.to_numeric(df['Reservoir Porosity (%)'], errors='coerce')
```

# Pressure Space

```python
# Reservoir Temperature and Pressure Space

# # Reservoir Temperature
df['Reservoir Temperature (C)'] = Ntemp + Ngeograd * (df['Mid-Point Reservoir Depth (m)'] - df['Water Depth(m)']) / 1000
# Hydrostatic Pressure (Mpa)
df['Hydrostatic Pressure (Mpa)'] = nsubhydpresgrad*(df['Mid-Point Reservoir Depth (m)'])/1000
# Lithostatic Pressure (Mpa)
df['Lithostatic Pressure (Mpa)'] = nlitpresgrad*(df['Mid-Point Reservoir Depth (m)'])/1000
# Matrix stress or effective stress coefficient (Zhang and Yin, 2017)
df['k0'] = 0.5+0.1/np.exp(3.28*df['Mid-Point Reservoir Depth (m)']/5100)
# Eaton Frac Pressure (Mpa)
# df['Eaton Frac Pressure (Mpa)'] = Npoisson/(1-Npoisson)*(df['Lithostatic Pressure (Mpa)']-df['Hydrostatic Pressure (Mpa)'])+df['Hydrostatic Pressure (Mpa)']
# Sand Frac Pressure (Mpa)
df['Sand Fracture Pressure (Mpa)'] = df['k0']*(df['Lithostatic Pressure (Mpa)']-df['Hydrostatic Pressure (Mpa)'])+df['Hydrostatic Pressure (Mpa)']
# Effective Stress (Mpa)
df['Effective Stress (Mpa)'] = df['Sand Fracture Pressure (Mpa)'] - df['Hydrostatic Pressure (Mpa)']
# Pressure Headroom (Mpa). Need to recheck f/ Alex calculation
df['Pressure Headroom (Mpa)'] = Nmaxp * df['Sand Fracture Pressure (Mpa)'] - df['Hydrostatic Pressure (Mpa)']
# Final Pressure (Mpa)
df['Final Pressure (Mpa)'] = df['Hydrostatic Pressure (Mpa)'] + df['Pressure Headroom (Mpa)']
# Delta Pressure (Mpa)
pdelta_mpa = df['Hydrostatic Pressure (Mpa)'] - df['Final Pressure (Mpa)']
```

# Compressibility

```python
# Total Compressibility
# Hydrostatic Pressure (psi)
df['Hydrostatic Pressure (psi)'] = df['Hydrostatic Pressure (Mpa)'] * 145
# Reservoir Temperature (F)
df['Reservoir Temperature (F)'] = df['Reservoir Temperature (C)']*9/5+32
# Reservoir Temperature (F)
df['Reservoir Temperature (K)'] = df['Reservoir Temperature (C)']+273
# Brine Compressibility (1/Mpa)
# df['Brine Compressibility'] = 1*145/(7.033*df['Hydrostatic Pressure (psi)']+541500*Nresbrine-537*df['Reservoir Temperature (F)']+403300)
# Brine Compressibility (1/Mpa) (Mathias et al., 2009)
df['Brine Compressibility'] = (10**-6 * (489 - 2 * df['Reservoir Temperature (C)'] +
                                         0.016 * df['Reservoir Temperature (C)']**2 - 
                                         1.3 * 10**-5 * df['Reservoir Temperature (C)']**3 - 
                                         0.666 * df['Hydrostatic Pressure (Mpa)'] - 
                                         0.004 * df['Reservoir Temperature (C)'] * df['Hydrostatic Pressure (Mpa)']) +
                               Nresbrine * (10**-6 * (300 - 2400 * Nresbrine + 
                                                     df['Reservoir Temperature (C)'] * (-13 + 47 * Nresbrine))))
# Rock Compressibility (1/Mpa)
df['Rock Compressibility'] = 0.00057
# Total Compressibility
df['Total Compressibility'] = df['Brine Compressibility'] + df['Rock Compressibility']
```

# CO2 Density, Storage Capacity, and Storage Efficiency

```python
# CO2 Density - Peng-Robinson Equation of State (PR EOS) (CoolProp)
# Temperature Reduction
tempred = df['Reservoir Temperature (K)']/crittemp
# Pressure Reduction
pred = df['Final Pressure (Mpa)']/critpressure

# Define fluid properties
fluid_name = 'CO2'  # Carbon dioxide

# Create an empty list to store the calculated densities
co2den = []

# Loop through the data points in the DataFrame
for index, row in df.iterrows():
    # Extract temperature and pressure from the DataFrame
    T_Kelvin = row['Reservoir Temperature (K)']
    P_MPa = row['Final Pressure (Mpa)']
    
    # Convert pressure to Pascal (CoolProp uses Pascal)
    P_Pascal = P_MPa * 1e6
    
    # Calculate density using CoolProp with Peng-Robinson EOS
    density = CP.PropsSI('D', 'T', T_Kelvin, 'P', P_Pascal, fluid_name)
    
    # Append the calculated density to the list
    co2den.append(density)
df["CO2 density (kg/m3)"] = co2den

# Add the list of densities to the DataFrame
co2den_capden = df['Reservoir Porosity (%)'] / 100 * df['Pressure Headroom (Mpa)'] * df['Total Compressibility'] * df["CO2 density (kg/m3)"]
df['CO2 Storage Capacity Density (kg/m3)'] = co2den_capden

# CO2 capacity density kg per m2
co2den_capden_m2 = co2den_capden * df['Net Reservoir Thickness (m)']
df['CO2 Storage Capacity Density (kg/m2)'] = co2den_capden_m2

# CO2 capacity density ton per km2
co2den_capden_km2 = co2den_capden_m2 * 1102.31
df['CO2 Storage Capacity (ton/km2)'] = co2den_capden_km2

# CO2 capacity density Megaton per km2
co2den_capden_mt_km2 = co2den_capden_km2 / 1e6
df['CO2 Storage Capacity (Mt/km2)'] = co2den_capden_mt_km2
```

```python
# CO2 average storage effeciency
co2_storeff =(df['Reservoir Porosity (%)'] * df['Total Compressibility'] * df['Effective Stress (Mpa)']) / df['Reservoir Porosity (%)'] * 100
df['CO2 Storage Average Efficiency'] = co2_storeff

# Pressure space density (MJ/m3)
pspaceden_mjm3 = df['Reservoir Porosity (%)'] / 100 * df['Effective Stress (Mpa)']
df['CO2 Storage Pressure Space Density (MJ/m3)'] = pspaceden_mjm3
```
